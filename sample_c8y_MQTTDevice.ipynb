```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

print("Welcome to the Jupyter data visualization demo!")
print(f"Device ID: {device_id}")
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue': [10500, 12600, 14800, 16200, 18900, 22400],
    'Expenses': [9200, 9600, 10100, 11200, 12500, 13800],
    'Customers': [120, 142, 158, 175, 190, 210]
}
df = pd.DataFrame(data)

print("Monthly Business Performance:")
display(df)
```

```python
df['Profit'] = df['Revenue'] - df['Expenses']

plt.figure(figsize=(10, 6))

plt.plot(df['Month'], df['Revenue'], 'o-', color='blue', label='Revenue')
plt.plot(df['Month'], df['Expenses'], 'o-', color='red', label='Expenses')
plt.plot(df['Month'], df['Profit'], 'o-', color='green', label='Profit')

plt.title('Revenue, Expenses, and Profit by Month')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
bars = plt.bar(df['Month'], df['Customers'], color='purple')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 5,
             f'{height}', ha='center', va='bottom')

plt.title('Monthly Customer Growth')
plt.xlabel('Month')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
